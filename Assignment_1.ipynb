##Exercise 1: Prime Numbers

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.


```python
def is_prime(n):
    # Prime numbers must be greater than 1
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False # Found a divisor, so it's not prime
    return True # No divisors found, so it is prime
```

## Exercise 2: Product of Random Numbers

Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


```python
import random

# Generate two random numbers
num1 = random.randint(1, 12)
num2 = random.randint(1, 12)
correct_product = num1 + num2

# Ask the user for the product
print(f"Time for a quick quiz!")
print(f"What is the sum of {num1} and {num2}?")

# Get and validate user input
try:
    user_answer = int(input("Your answer: "))

    # Check the answer
    if user_answer == correct_product:
        print(f" Correct, The answer is {correct_product}.")
    else:
        print(f" Incorrect, The correct answer is {correct_product}.")

except ValueError:
    print(" Invalid input. Please enter a valid number for your answer.")
```

##Exercise 3: Squares of Even/Odd Numbers


Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

```python
# I will calculate the squares of "EVEN" numbers.

start_range = 100
end_range = 200

print(f"-- Squares of EVEN Numbers in the range {start_range} to {end_range} --")

# We can start the range at 100 and step by 2 to only iterate over even numbers.
for number in range(start_range, end_range + 1, 2):
    square = number ** 2
    print(f"The square of {number} is {square}")
```

##Exercise 4: Hospital Billing System


A Hospital charges its patients based on the following:

* Room charges per day: ₹2,000

* Doctor consultation fee (flat): ₹1,500

* Lab test charges: Based on the number of tests taken. Each test costs ₹300.

* Medicine charges: Total cost of medicines provided.

* Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied.

Write a Python program to calculate the final bill for a patient who:

* Stayed for 4 days

* Had 3 lab tests

* Had medicine charges of ₹2,400



```python
def calculate_final_bill(days_stayed, num_lab_tests, medicine_cost):

   # -- Constants based on the Hospital's Policy --
    ROOM_RATE = 2000
    CONSULTATION_FEE = 1500
    LAB_TEST_COST_PER_UNIT = 300
    DISCOUNT_THRESHOLD = 10000
    DISCOUNT_PERCENTAGE = 0.10

    # 1. Calculate individual charges
    room_charges = days_stayed * ROOM_RATE
    lab_charges = num_lab_tests * LAB_TEST_COST_PER_UNIT

    # 2. Calculate Total Bill Before Discount
    total_bill_before_discount = (
        room_charges +
        CONSULTATION_FEE +
        lab_charges +
        medicine_cost
    )

    # 3. Apply Discount Rule
    discount_amount = 0.0
    if total_bill_before_discount > DISCOUNT_THRESHOLD:
        discount_amount = total_bill_before_discount * DISCOUNT_PERCENTAGE
        discount_message = f" Discount Applied (10%): -₹{discount_amount:,.2f}"
    else:
        discount_message = " No discount applied: Total bill is below ₹10,000 threshold."

    # 4. Calculate Final Bill
    final_bill = total_bill_before_discount - discount_amount

    # Output Detailed Summary
    print("      HOSPITAL BILLING SUMMARY     ")
    print("===================================")
    print(f"Patient Stay (Days): {days_stayed}")
    print(f"Lab Tests Taken:     {num_lab_tests}")
    print(f"Medicine Cost:       ₹{medicine_cost:,.2f}")
    print("-----------------------------------")
    print(f"Room Charges ({days_stayed}d * ₹{ROOM_RATE}):   ₹{room_charges:,.2f}")
    print(f"Doctor Consultation Fee:     ₹{CONSULTATION_FEE:,.2f}")
    print(f"Lab Charges ({num_lab_tests}t * ₹{LAB_TEST_COST_PER_UNIT}):     ₹{lab_charges:,.2f}")
    print(f"Medicine Charges:            ₹{medicine_cost:,.2f}")
    print("-----------------------------------")
    print(f"Total Bill (Before Discount): ₹{total_bill_before_discount:,.2f}")
    print(discount_message)
    print("===================================")
    print(f"FINAL BILL AMOUNT: **₹{final_bill:,.2f}**")
    print("===================================")

    return final_bill

# Patient Data from the Question
PATIENT_DAYS = 4
PATIENT_LAB_TESTS = 3
PATIENT_MED_COST = 2400

# Calculate and display the final bill
final_bill_result = calculate_final_bill(
    days_stayed=PATIENT_DAYS,
    num_lab_tests=PATIENT_LAB_TESTS,
    medicine_cost=PATIENT_MED_COST
)
```
