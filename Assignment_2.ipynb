1. Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Create a list of 5 integers
the_list = [10, 20, 30, 40, 50]
print(f"Original list: {the_list}")

# 1. append(element): Adds an element to the end of the list.
the_list.append(60)
print(f"After append(60): {the_list}")

# 2. extend(iterable): Appends the elements of an iterable (like another list).
the_list.extend([70, 80])
print(f"After extend([70, 80]): {the_list}")

# 3. insert: Inserts an element at a specific position.
the_list.insert(1, 15) # Insert 15 at index 1
print(f"After insert(1, 15): {the_list}")

# 4. remove: Removes the first occurrence of a specific value.
the_list.remove(40)
print(f"After remove(40): {the_list}")

# 5. pop: Removes and returns the element at a specific index (or the last if no index is given).
popped_element = the_list.pop(0) # Remove element at index 0.
print(f"Popped element (at index 0): {popped_element}")
print(f"After pop(0): {the_list}")

# 6. index: Returns the index of the first occurrence of a value.
index_of_30 = the_list.index(30)
print(f"Index of 30: {index_of_30}")

# 7. count: Returns the number of times a value appears in the list.
the_list.insert(3, 30) # Insert another 30 to test count
count_of_30 = the_list.count(30)
print(f"List with another 30: {the_list}")
print(f"Count of 30: {count_of_30}")

# 8. sort(): Sorts the list in-place (ascending by default)
the_list.sort()
print(f"After sort(): {the_list}")

# 9. reverse(): Reverses the order of elements in the list in-place
the_list.reverse()
print(f"After reverse(): {the_list}")

# 10. clear(): Removes all elements from the list
the_list.clear()
print(f"After clear(): {the_list}")
```

```python
# Create a tuple of 3 student names
stu_tuple = ('Mike', 'Hopper', 'Jane')
print(f"Original tuple: {stu_tuple}")

# Attempt to change the second name
try:
    stu_tuple[1] = 'Nancy'
except TypeError as e:
    print(f"Attempting to change the second name (stu_tuple[1] = 'Nancy') results in a:")
    print(f"Error: {e}")

# Explanation
print("Explanation:")
print("Tuples are 'immutable' data structures in Python.")
print("This means their contents 'cannot be changed' (modified, added, or removed) after they are created.")
print("The attempt to assign a new value to an element by its index results in a `TypeError: 'tuple' object does not support item assignment`.")
```

7. Given a number, check if it's positive, and if it is also even. If not positive, print if it's
zero or negative.

```python
# Define the number variable
number = -4

print(f"Checking number: {number}")

if number > 0:
    # Check if positive number is also even
    if number % 2 == 0:
        print("The number is 'Positive and Even'.")
    else:
        print("The number is 'Positive and Odd'.")
elif number == 0:
    # If not positive, check if it is zero
    print("The number is 'Zero'.")
else:
    # If not positive and not zero, it must be negative
    print("The number is 'Negative'.")
```

8. Create:

* A scalar using np.array(5)

* A 1D array with values 1 to 5

* A 2D array (2x3) with values from 10 to 60 in steps of 10




```python
import numpy as np

# A. Create a scalar
scalar = np.array(5)
print(f"A. Scalar:{scalar}")
print(f"   Shape: {scalar.shape}, Dimensions: {scalar.ndim}")

# B. A 1D array with values 1 to 5
array_1d = np.arange(1, 6)
print(f"B. 1D Array (1 to 5):{array_1d}")
print(f"   Shape: {array_1d.shape}, Dimensions: {array_1d.ndim}")

# C. A 2D array (2x3) with values from 10 to 60 in steps of 10
array_2d_flat = np.arange(10, 61, 10)
array_2d = array_2d_flat.reshape(2, 3)
print(f"C. 2D Array (2x3, 10 to 60 in steps of 10):{array_2d}")
print(f"   Shape: {array_2d.shape}, Dimensions: {array_2d.ndim}")
```

#Section C: Exploring Pandas

Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:

* Use .info(), .describe()

* Select 'Name' and 'City' columns

* Drop 'City' column

* Fill any missing values in 'Salary' column with the mean

* Remove any duplicate rows

```python
import pandas as pd
import numpy as np

# A. Create a small DataFrame manually with 10 rows
data = {
    'Name': ['Jems', 'Berlin', 'Sanu', 'Will', 'Raki', 'Arun', 'Jon', 'Pinkman', 'Walter', 'Fring'],
    'Age': [25, 30, 35, 22, 28, 40, 25, 32, 22, 28],
    'City': ['NY', 'SA', 'Chi', 'NY', 'LA', 'RSA', 'USA', 'NY', 'LA', 'CAN'],
    'Salary': [70000, 80000, 120000, 65000, np.nan, 90000, 70000, 85000, np.nan, 100000]
}
df = pd.DataFrame(data)
print("Initial DataFrame")
print(df)

# 1. Use .info()
print("\n 1. DataFrame Info")
df.info()

# 2. Use .describe()
print("\n 2. DataFrame Description")
print(df.describe())

# 3. Select 'Name' and 'City' columns
print("\n 3. Selected Columns ('Name', 'City')")
selected_cols = df[['Name', 'City']]
print(selected_cols)

# 4. Drop 'City' column
df.drop('City', axis=1, inplace=True)
print("\n 4. DataFrame after Dropping 'City' Column")
print(df)

# 5. Fill any missing values in 'Salary' column with the mean
salary_mean = df['Salary'].mean()
df['Salary'].fillna(salary_mean, inplace=True)
print(f"\n 5. DataFrame after Filling Missing 'Salary' (Mean: {salary_mean:.2f})")
print(df)

# 6. Remove any duplicate rows
df.drop_duplicates(inplace=True)
print("\n 6. DataFrame after Removing Duplicate Rows")
print(df)
```
