###Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set visualization style
sns.set_style('whitegrid')
```

```python
# Load the dataset
df = pd.read_csv('sales_data_with_discounts (2).csv')
```

```python
# Display the first few rows and column information to inspect the data
print(df.head())
```

```python
df.info()
```

###Descriptive Analytics for Numerical Columns

Objective: To compute and analyze basic statistical measures for numerical columns in the dataset.

```python
# Identify numerical columns
num_cols= ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
```

```python
print("Descriptive Statistics for Numerical Columns:")
df[num_cols].describe()
```

```python
# Calculating mean
print("mean")
mean_stats = df[num_cols].mean()
print(mean_stats.iloc[0])
```

```python
# Calculating mode
print("mode")
mode_stats = df[num_cols].mode()
print(mode_stats.iloc[0])
```

```python
# Calculating median
print("median")
median_stats = df[num_cols].median()
print(median_stats.iloc[0])
```

```python
# Calculating standard deviation
print("standard deviation")
std_stats = df[num_cols].std()
print(std_stats.iloc[0])
```

###Interpretation of Descriptive Statistics

Central Tendency: The Mean, Median (50th percentile), and Mode help identify the center point of the data distribution.

Dispersion: The Standard Deviation (std) shows the amount of variation or spread from the mean.

Shape: Comparing the mean and median can give an initial indication of skewness. The quartiles (25%, 50%, 75%) outline the distribution's shape.

##Data Visualization

##Histogram:

Plot histograms for each numerical column and analyze the distribution (e.g.
 skewness, presence of outliers).


```python
df[num_cols].hist(figsize=(12, 8), bins=20,color='orange')
plt.suptitle('Histograms of Numerical Columns', y=1.00)
plt.tight_layout()
plt.show()
```

INFERENCE:

* Outliers:  'Avg Price' and 'Total Sales Value' show significant outliers (extreme values) above the upper whisker, which is typical for sales data with high-value items.

* Spread:  The IQR (box length) and whiskers indicate the spread and variability of the middle 50% of the data.

* Distribution: The position of the median line within the box confirms the skewness observed in the histograms.

##Box Plots:

Create boxplots for numerical variables to identify outliers and the interquartile range.



```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(num_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=df[col],color='pink')
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.tight_layout()
```

Discuss any findings, such as extreme values or unusual distributions.

* Extreme Values:
The box plots for Avg Price, Total Sales Value, Discount Amount, and Net Sales Value are dominated by outliers.

  The "box" (representing the middle $50\%$ of all data, the IQR) is compressed at the very bottom, while hundreds of data points exist as individual outliers.

* Typical Range:

   For Net Sales Value, $50\%$ of all sales fall between $\$2,202.21$ (25th percentile) and $\$47,847.91$ (75th percentile).

   For Avg Price, $50\%$ of items are priced between $\$465$ and $\$10,100$. This wide range is again due to the mix of product categories (FMCG vs. Mobiles).
   
   Discount Rate (%) has a much "healthier" box plot, showing the main data distribution clearly with only a few outliers on the lower end.

##Bar Chart Analysis for Categorical Columns

```python
plt.figure(figsize=(18, 6))

# Business Unit Bar Chart
plt.subplot(1, 3, 1)
sns.countplot(data=df, x='BU', order=df['BU'].value_counts().index,color='pink')
plt.title('Frequency of Business Unit')
plt.xlabel('Business Unit')
plt.ylabel('Count')

# Day Bar Chart
plt.subplot(1, 3, 2)
sns.countplot(data=df, x='Day', order=df['Day'].value_counts().index)
plt.title('Frequency by Day of the Week')
plt.xlabel('Day')
plt.ylabel('Count')

# Brand Bar Chart
plt.subplot(1, 3, 3)
brands_to_plot = df['Brand'].value_counts().index
sns.countplot(data=df, x='Brand', order=brands_to_plot,color='gray')
plt.title('Frequency of Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.tight_layout()
```

###Analysis and Insights:

Business Unit (BU): The dataset is perfectly balanced across the three Business Units. Mobiles, FMCG, and Lifestyle each have exactly 150 records. This is excellent for comparing performance across categories.

Day: The sales records are also very evenly distributed across the days of the week, with each day having 60 to 70 records. There is no obvious "busiest" day based on the number of transactions.

Brand:

The Jeera and Viva (Lifestyle brands), and Babaji and Vedic (FMCG brands) are the most frequent.

The mobile brands (RealU, YouM, Sumsang, Orange) have fewer individual records but, as seen in the histograms, contribute to the highest sales values.

A summary of the key findings from the descriptive analytics and data visualizations .

1. Sales are Skewed: The dataset is characterized by a high volume of low-value transactions and a low volume of very high-value transactions. This heavily skews metrics like Total Sales Value and Avg Price.
