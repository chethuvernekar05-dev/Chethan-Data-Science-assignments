###6.	Fill missing BillAmount values with the mean bill amount.

```python
#Calculate the mean of the bill amount column.
mean_amt= cleaned_df['BillAmount'].mean()
```

```python
#Impute missing values in the bill amount column with the calculated mean.
impute_df = cleaned_df.copy()
impute_df['BillAmount'] = impute_df['BillAmount'].fillna(mean_amt)

print(f"Calculated Mean Bill Amount: {mean_amt:.2f}")
```
